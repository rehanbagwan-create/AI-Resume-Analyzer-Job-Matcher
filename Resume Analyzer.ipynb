```python
def extract_text_from_pdf(file):
    text = ""
    try:
        with open(file, "rb") as f:
            reader = PyPDF2.PdfReader(f)
            for page in reader.pages:
                if page.extract_text():
                    text += page.extract_text()
    except Exception as e:
        print("Error reading PDF:", e)
    return text
```
